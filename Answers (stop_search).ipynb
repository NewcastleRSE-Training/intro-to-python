## Exercises

### 1. Unique values

````python
# a. Create a list of the unique values that are in one of the columns.

unique_values = pandas.unique(df['location/street/name'])
print(unique_values)

# b. How many unique values are there in this column?
count = len(unique_values)
print(count)
````

### 2. Basic statistics


````python
# 2a. See what happens when you run the ```.describe()``` command with the 'officers_present' column and a non-numerical column.
print(df['officers_present'].describe())
````

````python
# non-numerical column (freq = most common value's frequency)
print(df['legislation'].describe())
````

````python
# 2b. What is the mean age of people stopped in this dataset?
print(df['age'].mean())
````

### 3. Grouping and aggregating

````python
# a. Print to the screen the mean ages of people in the dataset grouped by self_defined_ethnicity.
grouped_data = df.groupby('self_defined_ethnicity')
print(grouped_data.mean())
````

````python
# b. What is the max of numerical columns for stations grouped by outcome and gender?

grouped_data2 = df.groupby(['outcome', 'gender'])
print(grouped_data2.max())
````

### 4. Conditional selections

````python
# a.View only entries where the ethnicity is listed as 'White - English/Welsh/Scottish/Northern Irish/British'
# b. Save the results in a variable called `new_df` (new DataFrame) and view summary information to see how many entries 
# there are. 
new_df = df[df['self_defined_ethnicity'] == 'White - English/Welsh/Scottish/Northern Irish/British']
print(new_df.info())
````

````python
# c. Out of the entries in your new DataFrame, now narrow the selection to those that have are 35 years old or older.
print((new_df[new_df['age'] >= 35]).head(3))
````
